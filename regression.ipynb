```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cars.csv", sep=";")
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.select_dtypes(include=['int64', 'float64']).corr(), cmap="Blues", annot=True)
```

As we can see there is strong correlation between engine power and price

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=df, x="engine_power", y="units")
```

but still there are some outliers

```python
plt.figure(figsize=(8,12))
sns.boxplot(data=df, y="units")
```

```python
plt.figure(figsize=(8,12))
sns.boxplot(data=df, y="engine_power")
```
